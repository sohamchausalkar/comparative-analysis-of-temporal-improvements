## Normalize

```python
def normalize_difficulty(difficulty_series):
    difficulty_series = np.log1p(difficulty_series.replace(0, np.nan).fillna(1e-6))
    return (difficulty_series - difficulty_series.min()) / (difficulty_series.max() - difficulty_series.min()) + 0.1
```

```python
import glob 
ednet_difficulty = []
file_counter = 0
max_files = 500  
for file in glob.glob(ednet_path):
    if file_counter >= max_files:
        break
    user_df = pd.read_csv(file)
    difficulty_by_question = user_df.groupby('question_id')['elapsed_time'].mean()
    ednet_difficulty.append(difficulty_by_question)
    file_counter += 1
ednet_difficulty = pd.concat(ednet_difficulty)
ednet_difficulty = ednet_difficulty.groupby(ednet_difficulty.index).mean()
ednet_difficulty_normalized = normalize_difficulty(ednet_difficulty)
```

```python
assistments_df['start_time'] = pd.to_datetime(assistments_df['start_time'], format='mixed', errors='coerce')
assistments_df['end_time'] = pd.to_datetime(assistments_df['end_time'], format='mixed', errors='coerce')
assistments_df = assistments_df.dropna(subset=['start_time', 'end_time'])
assistments_df['elapsed_time'] = (assistments_df['end_time'] - assistments_df['start_time']).dt.total_seconds() * 1000
print("Elapsed Time Calculated Successfully.")
print(assistments_df[['start_time', 'end_time', 'elapsed_time']].head())
assistments_difficulty = assistments_df.groupby('problem_id')['elapsed_time'].mean()
assistments_difficulty_normalized = normalize_difficulty(assistments_difficulty)
```

## Temporal Improvements

```python

ednet_improvements = []
file_counter = 0

for file in glob.glob(ednet_path):
    if file_counter >= max_files:
        break
    user_id = os.path.basename(file).split('.')[0]
    user_df = pd.read_csv(file)
    question_groups = user_df.groupby('question_id')
    
    for question_id, group in question_groups:
        if len(group) > 1:
            initial_time = group.iloc[0]['elapsed_time']
            final_time = group.iloc[-1]['elapsed_time']
            improvement = (initial_time - final_time) / initial_time if initial_time > 0 else None
            if improvement == 0 or improvement is None:
                continue
            
            if question_id in ednet_difficulty_normalized:
                difficulty = ednet_difficulty_normalized[question_id]
                if difficulty > 0:  
                    normalized_improvement = improvement / difficulty
                    ednet_improvements.append({
                        'user_id': user_id,
                        'question_id': question_id,
                        'initial_time': initial_time,
                        'final_time': final_time,
                        'improvement': improvement,
                        'normalized_improvement': normalized_improvement
                    })
    file_counter += 1

ednet_df_improvements = pd.DataFrame(ednet_improvements)
```

```python
assistments_improvements = []
for (user_id, problem_id), group in assistments_df.groupby(['user_id', 'problem_id']):
    if len(group) > 1:
        initial_time = group.iloc[0]['elapsed_time']
        final_time = group.iloc[-1]['elapsed_time']
        improvement = (initial_time - final_time) / initial_time if initial_time > 0 else None
        if improvement == 0 or improvement is None:
            continue
        
        if problem_id in assistments_difficulty_normalized:
            difficulty = assistments_difficulty_normalized[problem_id]
            if difficulty > 0:  
                normalized_improvement = improvement / difficulty
                assistments_improvements.append({
                    'user_id': user_id,
                    'problem_id': problem_id,
                    'initial_time': initial_time,
                    'final_time': final_time,
                    'improvement': improvement,
                    'normalized_improvement': normalized_improvement
                })
assistments_df_improvements = pd.DataFrame(assistments_improvements)
```

## Data Cleaning

```python
problematic_problems = [41497, 545058, 57389, 610349, 12347]

assistments_df_improvements_filtered = assistments_df_improvements[
    ~assistments_df_improvements['problem_id'].isin(problematic_problems)
]

assistments_df_improvements_filtered['improvement'] = assistments_df_improvements_filtered['improvement'].clip(-10, 10)
assistments_df_improvements_filtered['initial_time'] = assistments_df_improvements_filtered['initial_time'].clip(1000, 300000)
assistments_df_improvements_filtered['final_time'] = assistments_df_improvements_filtered['final_time'].clip(1000, 300000)

def robust_normalize(difficulty_series):
    difficulty_series = difficulty_series.replace(0, np.nan).fillna(1e-6)
    median = difficulty_series.median()
    mad = np.median(np.abs(difficulty_series - median))
    return (difficulty_series - median) / (mad if mad > 0 else 1)

assistments_difficulty_normalized = robust_normalize(assistments_difficulty)


assistments_difficulty_clipped = assistments_difficulty.clip(lower=1, upper=assistments_difficulty.quantile(0.99))
assistments_difficulty_normalized = np.log1p(assistments_difficulty_clipped)
assistments_difficulty_normalized = (assistments_difficulty_normalized - assistments_difficulty_normalized.min()) / (
    assistments_difficulty_normalized.max() - assistments_difficulty_normalized.min()
)

assistments_df_improvements_filtered['normalized_improvement'] = (
    assistments_df_improvements_filtered['improvement'] /
    assistments_difficulty_normalized[assistments_df_improvements_filtered['problem_id']].values
)
```

```python
from scipy.stats import gamma, lognorm, weibull_min
from scipy.stats import chisquare, norm
import pandas as pd
import matplotlib.pyplot as plt

def chi_square_test(data, distribution, dist_name, bins=10):
    observed_frequencies, bin_edges = np.histogram(data, bins=bins)
    total_data_points = len(data)

    if distribution == "gamma":
        params = gamma.fit(data)  
        expected_frequencies = total_data_points * (gamma.cdf(bin_edges[1:], *params) - gamma.cdf(bin_edges[:-1], *params))
    elif distribution == "lognorm":
        params = lognorm.fit(data) 
        expected_frequencies = total_data_points * (lognorm.cdf(bin_edges[1:], *params) - lognorm.cdf(bin_edges[:-1], *params))
    elif distribution == "weibull":
        params = weibull_min.fit(data)  
        expected_frequencies = total_data_points * (weibull_min.cdf(bin_edges[1:], *params) - weibull_min.cdf(bin_edges[:-1], *params))
    else:
        raise ValueError("Unsupported distribution type")

    expected_frequencies *= observed_frequencies.sum() / expected_frequencies.sum()
    chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

    return chi2_stat, p_value


ednet_data = ednet_improvements_clipped['normalized_improvement']
assistments_data = assistments_df_improvements_filtered['normalized_improvement']
results = []
for dataset_name, data in [("EdNet", ednet_data), ("ASSISTments", assistments_data)]:
    for dist_name in ["gamma", "lognorm", "weibull"]:
        chi2_stat, p_value = chi_square_test(data, dist_name, dist_name)
        results.append({
            "Dataset": dataset_name,
            "Distribution Tested": dist_name.capitalize(),
            "Chi-Square Statistic": round(chi2_stat, 2),
            "p-value": f"{p_value:.4E}" if p_value > 0 else "< 0.0001",
            "Conclusion": "Reject: Does not follow" if p_value < 0.05 else "Fail to reject: Follows"
        })

updated_chi_square_results = pd.DataFrame(results)

fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('tight')
ax.axis('off')
table = ax.table(
    cellText=updated_chi_square_results.values,
    colLabels=updated_chi_square_results.columns,
    cellLoc='center',
    loc='center'
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width(col=list(range(len(updated_chi_square_results.columns))))
plt.title("Chi-Square Goodness-of-Fit Test Results for Various Distributions", fontsize=14, weight='bold', pad=20)
plt.show()

print(updated_chi_square_results)
```
